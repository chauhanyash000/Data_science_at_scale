```python
data = pd.read_csv('file:///Users/yashchauhan/Desktop/sanfrancisco_incidents_summer_2014.csv')
counts = data['DayOfWeek'].value_counts()
cc_counts = data['Category'].value_counts()
```

## Analysis and Conclusions ##

1. Larceny/Theft is the most frequenctly occuring crime in the city with more than 8000 criminal cases filed for the category
2. After Larceny/Theft other offences, non criminal and assault category crimes are taking place with little difference in the numbers of cases filed under each category
3. Most Crimes are happening on Friday which is the last working day of week but comparatively crimes occur at all the days with little but noticeble difference of more crimes occuring on weekends than in weekdays
4. Larceny/Theft and Assault are taking place more on Friday, Saturday, Sunday
5. In contrast to above categories, Drug/Narcotics and other offences are taking place more during weekdays.
